```python
student_name = input("Enter student's name:")
score_input = input("Enter student's score:")

try:
    score = int(score_input)
    if score <= 90:
        grade = 'A'
    elif score >= 80 < 90:
        grade = 'B'
    elif score >= 70 < 80:
        grade = 'C'
    elif score < 70:
        grade = 'F'
    else:
        grade = 'Invalid Score (out of range)'

    print(f"\nStudent Name: {student_name}")
    print(f"Score: {score}")
    print(f"Letter Grade: {grade}"))

```
